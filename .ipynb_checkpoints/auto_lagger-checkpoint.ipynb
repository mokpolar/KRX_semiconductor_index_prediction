```python
import pandas as pd
```

```python
df = pd.read_csv('total_data.csv')
```

```python
df.head()
```

```python
df = df[['date', 'KRXsemiconductor_close', 'SKhynix_close']]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace = True)
```

```python
# df should involve datetime index
# target column example = 'SKhynix_close'
def auto_lagger(df, target_column):
    # lagging day list
    lag_days = [1, 7, 30, 90, 180, 365]
    col_name =  []
    # creating new column name
    for i in lag_days:
        col_name.append(target_column + str(i))
    # appending lagged columns
    for i, j in enumerate(lag_days):
        df[locals()['col_name'][i]] = df[target_column].shift(periods = j, axis = 0)
    return df
```

```python
auto_lagger(df, 'SKhynix_close')
```
