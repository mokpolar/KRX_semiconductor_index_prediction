```python
def make_change(x):
    output = []
    for i in x:
        if i == x[-1]:
            print('end')
        else:
            change = x[i+1] / x[i]
            output.appned(change)
    return output      
        
```

```python
test = df_raw['krx_close'].values
```

```python
test[-1]
```

```python
def make_change(x):
    output = ['NAN']
    for i, j in enumerate(x):
        if i == len(x)-1:
            break
        else:
            output.append((x[i+1] / x[i]))
    return(output)
```

```python
test_2 = make_change(test)
```

```python
df_raw2['krx_change'] = pd.Series(test_2)
```
