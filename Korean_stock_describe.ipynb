```python
KOSPI200 = pd.read_excel('Data\\KOSPI200.xlsx', sheet_name='Sheet1')
KRXsemi = pd.read_excel('Data\\KRX반도체.xlsx', sheet_name='Sheet1')
SKHynix = pd.read_excel('Data\\SK하이닉스.xlsx', sheet_name='Sheet1')
SamElec = pd.read_excel('Data\\삼성전자.xlsx', sheet_name='Sheet1')
```

```python
describe = KOSPI200.describe()
KRXsemi_describe = KRXsemi.describe()
SKHynix_describe = SKHynix.describe()
SamElec_describe = SamElec.describe()
```

```python
KOSPI200_date = pd.DataFrame([[KOSPI200['DATE'].min(), KOSPI200['DATE'].max(), len(KOSPI200['DATE']), 'KOSPI200', np.nan, np.nan]],
                             columns=['OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME', 'CHANGE'])
KRXsemi_date = pd.DataFrame([[KRXsemi['DATE'].min(), KRXsemi['DATE'].max(), len(KRXsemi['DATE']), 'KRXsemi', np.nan, np.nan]],
                             columns=['OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME', 'CHANGE'])
SKHynix_date = pd.DataFrame([[SKHynix['DATE'].min(), SKHynix['DATE'].max(), len(SKHynix['DATE']), 'SKHynix', np.nan, np.nan]],
                             columns=['OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME', 'CHANGE'])
SamElec_date = pd.DataFrame([[SamElec['DATE'].min(), SamElec['DATE'].max(), len(SamElec['DATE']), 'SamElec', np.nan, np.nan]],
                             columns=['OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME', 'CHANGE'])
```

```python
SamElec_date
```

```python
describe = pd.concat([describe, KOSPI200_date], axis=0)
describe = pd.concat([describe, KRXsemi_describe], axis=0, sort=False)
describe = pd.concat([describe, KRXsemi_date], axis=0, sort=False)
```

```python
describe
```

```python
describe = pd.concat([describe, SKHynix_describe], axis=0, sort=False)
describe = pd.concat([describe, SKHynix_date], axis=0, sort=False)
```

```python
describe
```

```python
describe = pd.concat([describe, SamElec_describe], axis=0, sort=False)
describe = pd.concat([describe, SamElec_date], axis=0, sort=False)
```

```python
describe.to_csv('describe.csv')
```
