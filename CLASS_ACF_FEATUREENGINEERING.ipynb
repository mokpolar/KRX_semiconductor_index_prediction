```python
'''
# Feature Engineering Type

1. return lagger : 이미 생성                                                                             
2. 차분 : MoM, QoQ, YoY 등의 유형은 interpolation 전에 그냥 서로 빼면 될듯
3. 그대로 쓰는 input

---
# Feature Engineering

그러면 일단 input은 dataframe이 되고
function 을 저 세 종류를 짜고
그래고 class 하나에 몰아서 
column별로 순회하면서 
key값 기준으로 판단하고
1 번 유형이면 return function으로 늘이고
2 번 유형이면 그냥 빼고 
3 번 유형이면 놔두고
---
# Interpolation

arg parser로 입력 값으로 해봐도 재밌겠군

일단 전체 순회하면서 늘린 column 생성 한 이후에 interpolation 하는게 맞을 듯 한데 

유형은 3가지로 
step, linear, spline
---
그리고 빈 공간이 하나도 없는 데이터프레임을 뱉으면 된다. 
---
'''  
```
